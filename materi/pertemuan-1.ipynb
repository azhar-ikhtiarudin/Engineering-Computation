```python
x=np.linspace(-np.pi, np.pi, 100)
y=np.sin(x)
y1=np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,y1)
plt.title('title sample')
plt.grid(True)
matplotlib.style.use('seaborn')
```
