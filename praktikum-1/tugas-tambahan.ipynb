<h5>Single Precision forward </h5>

```python
import numpy as np
def f(x):
    return np.float32(1)/np.float32(x**np.float32(4))

# n =np.float32(10000)
approx_value=np.float32(0)
true_value = np.float32(np.pi)**np.float32(4)/np.float32(90)
print("      i                  f(i)")
print("--------------------------------------------------------")

for i in range(10000):
    approx_value += f(i+1)
    print("%18.10f %18.10f " % (i+1, approx_value))
    
print("   Approx Value:   %18.10f            " % (approx_value))
print("   True Value:     %18.10f            " % (true_value))
print("   Approx Value:   ", (type(approx_value)))
print("   True Value:     ", (type(true_value)))
print("   type f(10000):  ", (type(approx_value)))
```

<h3><mark style="background-color: lightskyblue">SOAL 13</mark></h2>

```python
import math

def f(x, i):
    return (-1)**i*x**(2*i)/math.factorial(2*i)

approx_value = 0

angle = math.pi/3

true_value = math.cos(angle)

print("     i           f(i)                error")
print("     --------------------------------------------")

for i in range(10):
    approx_value += f(angle, i)
    error = abs((approx_value - true_value)/true_value) 
    print("     %i     %18.10f  %18.10f " % (i, approx_value, error))
    
print("     --------------------------------------------")
print("approx_value:    %18.10f " % (approx_value))
print("true_value:      %18.10f " % (true_value))
print("absolute error:  %18.10f" % (error))
```
